We are going to take datasets from 3 main files and put them into pandas dataframe

```python
# read the dataset
df_meanTemp: pd.DataFrame = pd.read_csv('./data/Monthly_Mean_Temp.csv')
df_incomeByTown: pd.DataFrame = pd.read_csv('./data/Income_By_Town.csv')
df_electricConsumeByTown = pd.DataFrame = pd.read_csv('./data/Electric_Consumption_By_Town.csv')
```

```python
# In this dataset we are interested in obtaining the mean air temperature. Later on, we will be getting the median value in between the max and the min
df_meanTemp = df_meanTemp[
    df_meanTemp["Data Series"].isin([
        "Air Temperature Means Daily Maximum (Degree Celsius)",
        "Air Temperature Means Daily Minimum (Degree Celsius)"
    ])
].copy()
```

<div class="alert alert-block alert-info">
<b>Observation:</b> Notice that the way how the date represented in `df_electricConsumeByTown` differs from how it is represented in `df_meanTemp`.<br>
Later on we will standardise it and make them into the same row/column in case we intend to merge them together.
</div>

```python
# transpose the dataframe to let the date period be represented at the top instead
df_electricConsumeByTown = df_electricConsumeByTown.T

df_electricConsumeByTown.columns = df_electricConsumeByTown.iloc[0]
df_electricConsumeByTown = df_electricConsumeByTown.drop(df_electricConsumeByTown.index[0])

# convert the date from '2025-01' to '2025 Jan' so that it is easier to merge the dataframe together later on
def changeDaDate(col: str) -> str:
    month_abbr = ['', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

    if '-' in col:
        parts = col.split('-')
        if len(parts) == 2 and parts[1].isdigit():
            year, month_num = parts
            month_num_int = int(month_num)
            if 1 <= month_num_int <= 12:
                return f"{year} {month_abbr[month_num_int]}"
    return col

df_electricConsumeByTown.columns = [changeDaDate(col) for col in df_electricConsumeByTown.columns]
```

```python
# check to see if it tranforms properly
df_electricConsumeByTown.head(3)
```

<div class="alert alert-block alert-info">
<b>Observation:</b> Notice that the in the `df_electricConsumeByTown` dataframe, the earliest data originates from 2005 Jan, whereas the `df_meanTemp` dataframe has datasets from 1975.<br>
We would want to standardize and get the maximum time frame that we can get from both dataframe to make it more homogenous.
</div>

```python
# identify date columns 
identifierCols = ['Data Series']

# notice that there is a trailing space at the back of each dateMeanTemp, so we need to remove before we can compare the intersect.
dateMeanTemp = [col for col in df_meanTemp.columns if col not in identifierCols]
newDateMeanTemp = {col: col.strip() for col in dateMeanTemp}
df_meanTemp = df_meanTemp.rename(columns=newDateMeanTemp)

# get the date columns from the dataframe respectively
dateElecConsumeTown = set(df_electricConsumeByTown.columns)
dateMeanTemp = set(df_meanTemp.columns) - set(identifierCols)

# find the common columns between the 2 dataframes
commonDates = dateElecConsumeTown.intersection(dateMeanTemp)

# for df_electricConsumeByTown, keep only the common columns
df_electricConsumeByTown = df_electricConsumeByTown.loc[:, sorted(commonDates)]

# for df_meanTemp, keep the ['Date Series'] + common columns
colsToKeepMeanTemp = ['Data Series'] + sorted(commonDates)
df_meanTemp = df_meanTemp.loc[:, colsToKeepMeanTemp]
```

```python
# we would like to sort the data according to ascending date order
monthStrToNum = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6, 'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}

def sortableDate(col):
    try:
        year, mon = col.split()
        return (int(year), monthStrToNum[mon])
    except Exception:
        return (9999, 99) #put wrong columns at the back

# now we sort
commonDatesSorted = sorted(commonDates, key=sortableDate)

df_electricConsumeByTown = df_electricConsumeByTown.loc[:, commonDatesSorted]
df_meanTemp = df_meanTemp.loc[:, ['Data Series'] + commonDatesSorted]
```
